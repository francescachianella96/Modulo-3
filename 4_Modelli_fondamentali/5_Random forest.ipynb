```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestRegressor
```

Random forest in python

```python
cali = fetch_california_housing()
X, y = cali.data, cali.target
features_names = cali.feature_names
```

```python
rf_regressor = RandomForestRegressor(
    n_estimators=100, 
    max_depth=20,
    random_state=42
)

rf_regressor.fit(X, y)
```

```python
# verifichiamo l'importanza delle feature
importances = rf_regressor.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
# andiamo a verificare con graficoo a barre l'importanza delle feature
plt.figure(figsize=(10,6))
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), [features_names[i] for i in indices], rotation=45)
plt.tight_layout()
plt.grid()
plt.show()
```

```python
print ("Importanza feature")
for i in range(5):
    print(f"{i +1} {features_names[indices[i]]}: {importances[indices[i]]:.4f}")
```

```python
# grafico per vedere il primo albero
tree_num = 0
max_depth = 3
estimator = rf_regressor.estimators_[tree_num]
print(f"Total estimator:\t{len(rf_regressor.estimators_)}")
plt.figure(figsize=(20,10))
plot_tree(
    estimator, 
    feature_names=features_names, 
    filled=True, 
    max_depth=max_depth, 
    fontsize=10
)
plt.title(f"Decision tree #{tree_num +1} in Random Forest \n$max\\_depth = {max_depth}$")
```

```python
# grafico per vedere l'ultimo albero
tree_num = -1
max_depth = 3
estimator = rf_regressor.estimators_[tree_num]
print(f"Total estimator:\t{len(rf_regressor.estimators_)}")
plt.figure(figsize=(20,10))
plot_tree(
    estimator, 
    feature_names=features_names, 
    filled=True, 
    max_depth=max_depth, 
    fontsize=10
)
plt.title(f"Decision tree #{tree_num +1} in Random Forest \n$max\\_depth = {max_depth}$")
```

```python
y_pred = rf_regressor.predict(X)

```

```python
plt.figure(figsize=(8,6))
plt.scatter(y, y_pred, alpha=0.3)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Real")
plt.ylabel("Predicted")
plt.title("Random Forest Regression California Housing")
plt.tight_layout()
plt.show()
```

```python
err = np.abs(y - y_pred)

tol = 0.2 * y.mean()
good_pred = np.sum(err < tol)
out = good_pred / len(y) *100

print(f"Prediction within 20% from average:\t{out:.2f}%")
```

```python
mean_err = np.mean(err/y)
print(f"Mean error:\t{mean_err *100 :.2f}%")
```
