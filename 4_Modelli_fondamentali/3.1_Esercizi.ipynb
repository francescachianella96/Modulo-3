Decision tree in python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, plot_tree
```

# Classificazione - dataset wine

```python
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
print(X_iris[:5])
print(y_iris[:5])
```

```python
df_classifier = DecisionTreeClassifier(
    max_depth=3, 
    random_state=42
)

dt_classifier = df_classifier.fit(X_iris, y_iris)
y_pred_class = dt_classifier.predict(X_iris)
```

```python
# verifica albero e numero foglie
print(f"DT Depth: {dt_classifier.get_depth()}")
print(f"DT # Leaves: {dt_classifier.get_n_leaves()}")
```

```python
plt.figure(figsize=(14, 10))
plot_tree(
    dt_classifier, 
    feature_names=iris.feature_names, 
    class_names=iris.target_names, 
    filled=True,
    fontsize=10
)
plt.title("Decision Tree Classifier - Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(range(len(y_iris)), y_iris, label="Real", alpha=0.6)
plt.scatter(range(len(y_iris)), y_pred_class, label="Predicted", alpha=0.4)
plt.title("Decision Tree Classifier")
plt.xlabel("Samples")
plt.ylabel("Classes")
plt.legend()
plt.show()
```

```python
classification_accuracy = (y_pred_class == y_iris).sum() / len(y_iris)
print(f"Classification Accuracy: {classification_accuracy*100:.2f}%")
```
