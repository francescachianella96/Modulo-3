# Precision

```python
from sklearn.metrics import precision_score
```

```python
y_true_bin = ["A"] * 12 + ["B"] * 8
y_pred_bin = ["A"] * 10 + ["B"] * 8 + ["A"] * 2
```

```python
precision_A = precision_score(y_true_bin, y_pred_bin, pos_label="A")
precision_B = precision_score(y_true_bin, y_pred_bin, pos_label="B")
print(f"Precision (A):\t{precision_A *100:.2f}%\nPrecision (B):\t{precision_B *100:.2f}%")
```

```python
y_true = ["A", "A", "A", "B", "B", "B", "C", "C", "C"]
y_pred = ["A", "B", "A", "B", "B", "C", "C", "B", "C"]
```

```python
precision_micro = precision_score(y_true, y_pred, average="micro")
precision_macro = precision_score(y_true, y_pred, average="macro")
precision_weighted = precision_score(y_true, y_pred, average="weighted")
print(f"Precision micro:\t{precision_micro *100:.2f}%")
print(f"Precision macro:\t{precision_macro *100:.2f}%")
print(f"Precision weighted:\t{precision_weighted *100:.2f}%")
```
