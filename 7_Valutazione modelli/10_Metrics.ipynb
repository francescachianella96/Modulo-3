# Metrics

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, roc_curve
```

```python
# importazione dataset

data = load_breast_cancer()
X = data.data
y = data.target

df = pd.DataFrame(X, columns=data.feature_names)
df["target"] = y
display(df)
```

```python
# data splitting
# con stratify=y manteniamo la stessa proporzione delle classi sia nel train che nel test set

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    shuffle=True,
    stratify=y,
    test_size=0.2,
    random_state=42
)
```

```python
# scaling dei dati

scaler = StandardScaler()
X_train_scaled = X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# importazione del modello e addestramento

model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
# generazione predizioni

y_pred = model.predict(X_test_scaled)
y_prob = model.predict_proba(X_test_scaled)[:,1]
```

```python
# calcolo metriche

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred)
```

```python
# print dei parametri

print(f"Accuracy:\t{accuracy*100:.2f}%")
print(f"Precision:\t{precision*100:.2f}%")
print(f"Recall:\t\t{recall*100:.2f}%")
print(f"F1-score:\t{f1*100:.2f}%")
print(f"AUC:\t\t{auc:.3f}")
```

```python
# rappresentazione curva roc

fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, alpha=0.3, label=f"AUC = {auc:.3f}")
plt.fill_between(fpr, tpr, alpha=0.3, color="gray")
plt.plot([0,1], [0,1], linestyle="--")                  # plot della diagonale da 0,0 a 0,1
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("AUC ROC Logistic Regression")
plt.legend(loc="center right")
plt.grid(True)
plt.show()
```
