# Recall

```python
from sklearn.metrics import recall_score
```

```python
y_true_bin = ["A"] * 12 + ["B"] * 8
y_pred_bin = ["A"] * 10 + ["B"] * 8 + ["A"] * 2
```

```python
recall_A = recall_score(y_true_bin, y_pred_bin, pos_label="A")
recall_B = recall_score(y_true_bin, y_pred_bin, pos_label="B")
print(f"Recall (A):\t{recall_A *100:.2f}%\nRecall (B):\t{recall_B *100:.2f}%")
```

```python
y_true = ["A", "A", "A", "B", "B", "B", "C", "C", "C"]
y_pred = ["A", "B", "A", "B", "B", "C", "C", "B", "C"]
```

```python
recall_micro = recall_score(y_true, y_pred, average="micro")
recall_macro = recall_score(y_true, y_pred, average="macro")
recall_weighted = recall_score(y_true, y_pred, average="weighted")
print(f"Recall micro:\t\t{recall_micro *100:.2f}%")
print(f"Recall macro:\t\t{recall_macro *100:.2f}%")
print(f"Recall weighted:\t{recall_weighted *100:.2f}%")
```
