```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, multilabel_confusion_matrix
```

```python
y_true_bin = [0, 1, 0, 1, 0, 1, 1]
y_pred_bin = [0, 1, 0, 0, 0, 1, 1]

y_true_multi = [0, 1, 2, 1, 0, 2, 1]
y_pred_multi = [0, 2, 2, 1, 0, 0, 1]
```

```python
cm_bin = confusion_matrix(y_true_bin, y_pred_bin)
print(f"Confusion matrix (binary):\n{cm_bin}")
```

```python
cm_multi = multilabel_confusion_matrix(y_true_multi, y_pred_multi)
print(f"Multilabel Confusion matrix:\n{cm_multi}")
```

```python
# rappresentiamola graficamente con la heatmap (binaria)

classes_bin = sorted(set(y_true_bin))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_bin, annot=True, fmt="d", cmap="Blues",
            xticklabels = classes_bin,
            yticklabels = classes_bin)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# rappresentiamola graficamente con la heatmap (multiclasse)

cm_multi = confusion_matrix(y_true_multi, y_pred_multi)

classes_multi = sorted(set(y_true_multi))
plt.figure(figsize=(8, 6))
sns.heatmap(cm_multi, annot=True, fmt="d", cmap="Blues",
            xticklabels = classes_multi,
            yticklabels = classes_multi)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
